```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import anthropic
import requests

load_dotenv()
OpenAIModels=["gpt-4.1","gpt-4.1-mini","gpt-4o","gpt-5-mini","gpt-5.2-chat","gpt-audio-mini"]
AnthropicModels=["claude-opus-4-6","claude-sonnet-4-6"]

client = OpenAI(
    api_key=os.getenv("AZURE_API_KEY"),
    base_url=os.getenv("AZURE_BASE_URL").rstrip("/") + "/openai/v1",
    default_headers={"api-key": os.getenv("AZURE_API_KEY")},
)

response = client.chat.completions.create(
    model=OpenAIModels[0],  # or "gpt-4.1-mini" or "gpt-4o"
    messages=[
        {"role": "user",
         "content": "say Hi"}
    ]
)
print("GPT Output:")
print(response.choices[0].message.content)

#alternate method:
url = os.getenv("AZURE_BASE_URL").rstrip("/") + "/openai/v1/responses"
headers = {"api-key": os.getenv("AZURE_API_KEY"), "Content-Type": "application/json"}
payload = {
    "model": OpenAIModels[0],
    "input": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Write a short poem about the ocean."},
    ],
    "temperature": 1,
    "top_p": 1,
    "stream": False,
    "store": False,
}
r = requests.post(url, headers=headers, json=payload, timeout=60)
# print(r.status_code)
# print(r.text)
r.raise_for_status()
print("GPT Output2:")
print(r.json()["output"][0]["content"][0]["text"])


client2 = anthropic.Anthropic(
    api_key=os.getenv("AZURE_API_KEY"),
    base_url=os.getenv("AZURE_BASE_URL").rstrip("/") + "/anthropic",
)

response = client2.messages.create(
    model="claude-opus-4-6",  # or "claude-sonnet-4-6"
    max_tokens=1024,
    messages=[
        {"role": "user",
         "content": "Say hello to me."}
    ]
)
print("Anthropic Output:")
print(response.content[0].text)
```
