```python
import os
import requests
from dotenv import load_dotenv
from openai import AzureOpenAI

load_dotenv() # reads.env file into environment variables
base = os.getenv("AZURE_BASE_URL").rstrip("/")
key = os.getenv("AZURE_API_KEY")
headers = {
    "api-key": key,
    "Ocp-Apim-Subscription-Key": key,
}

paths = [
    "/",
    "/openai",
    "/openai/models",
    "/openai/models?api-version=2024-10-21",
    "/openai/deployments",
    "/openai/deployments?api-version=2024-10-21",
    "/openai/deployments/gpt-4o-mini/chat/completions?api-version=2024-10-21",
    "/v1/models",
    "/v1/chat/completions",
]

for path in paths:
    url = base + path
    r = requests.get(url, headers=headers)
    status = r.status_code
    body = r.text[:150]
    print(f"{status}  {path}")
    print(f"     {body}\n")

# client = AzureOpenAI(api_key=os.getenv("AZURE_API_KEY"),
#                     azure_endpoint=os.getenv("AZURE_BASE_URL"),
#                     api_version=os.getenv("AZURE_API_VERSION")
#                      )
# response = client.chat.completions.create(
#     model="gpt-4o-mini",
#     messages=[
#         {"role": "user", "content": "Explain why finetuning can improve task-specific performance."}
#     ]
# )
#
# print(response.choices[0].message.content)
```
